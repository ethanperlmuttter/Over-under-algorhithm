```python
#returns suggested base amount for bet
#over/unders are generally in slight minus odds, for the purpose of this excercise we will assume a line of -112
#This means a $112 dollar bet pays out $100
#For the example above this function should return 112, NOT 100
#If using this function for purposes other than this excercise, all values can be scaled by a constant value to fit
#your personal financial constraints
#returns 0 if prediction and projection are within threshold
def wager(thresh, base, rate, pred, proj):
    value = 0
    thresh2 = thresh
    if pred<=proj:
        thresh2 = -1*thresh
        
    if abs(pred-proj)<=thresh:
        return value
    else:
        value = base + rate*abs(pred-proj-thresh2)
        return value


        
        
```

```python
wager(2, 1, 1, 200, 199)
```

```python
print(wager(2, 1, 1, 200, 201))
print(wager(2,1,1, 200, 203))
print(wager(2, 1, 1, 200, 199))
print(wager(2,1,1, 200, 197))
```

```python
#iterates through seeding games and simulates wagering strategy

def test_strat(init, thresh, base, rate):
    print("Initial Balance: $",init)
    balance = init
    staked = 0
    vig = 0.8928571429
    wagered = 0
    wins = 0
    for i in range(84):
        for j in range(1,4):
            final = df2.iloc[i]["Final Total"]
            model = df2.iloc[i]["Q"+str(j)+" Prediction"]
            line = df2.iloc[i]["Q"+str(j)+" Projected Score"]
            wag = wager(thresh, base, rate, model, line)
            wagered = wagered + (wag!=0)
            if final == line:
                balance = balance +0
            elif (model<line) == (final<line):
                balance = balance + vig*wag
                wins = wins + (wag!=0)
            else:
                balance = balance - wag
            staked = staked + wag
    roi = 1.0*(balance-init)/staked
    avg_wag = staked/wagered
    win_rate = wins/wagered
    print("Final Balance: $",balance)
    print("Total Staked: $",staked)
    print("ROI: ", roi)
    print("Average wager: $", avg_wag)
    print("Win Rate: ", win_rate)
    
    
   
    

                
```
